# Group by

```python
import pandas as pd
df = pd.read_csv("weather.csv")
df
```

```python
for name, df in df.groupby('city'):
    print('City is:', name)
    print('Maximum tempriture is: ', df['temperature'].max())
    print('\n')
    print(df)
    print('\n')
    
```

```python
for name, df in df.groupby(['city', 'event']):
    print('Group is:', name)
    print('Maximum tempriture is: ', df['temperature'].max())
    print('\n')
    print(df)
    print('\n')
```

### Passing a function in group by

```python
def get_letter_type(letter):
       if letter.lower()[0] in 'dct':
        return 'DCT'
       else:
        return 'Other'
```

```python
for group, frame in df.groupby(get_letter_type, axis=1):
    print(group)
    print("\n")
    print("\n")
    print("\n")
    print(frame)
```
